```python
from keras.models import load_model
from PIL import Image
import numpy as np
from astropy.io.fits import getdata, PrimaryHDU
from matplotlib import pyplot as plt
```

```python
model = load_model('crop_checkpoints/checkpoint-16.h5')
```

```python
def get_png_data(filename):
    """
    Get the PNG data for the given FITS file
    
    Args:
        filename: Path to the FITS file
    """
    data = getdata(filename)
    data = data.byteswap().newbyteorder()
    
    file = Image.fromarray(data)
    file = file.convert('L')  # black and white
    file.save('test.png')
    
    png_data = Image.open('test.png')
    png_data = np.array(png_data)
    
    return png_data
```

```python
png = get_png_data('M13V20120711.fits')
```

```python
pred = model.predict(png.reshape(*(png.shape), 1, 1))
```

```python
pred.shape
```

```python
plt.imshow(pred.reshape(951, 951), cmap='gray');
```

```python
img = Image.fromarray(pred.reshape(951, 951))
img = img.convert('L')
img.save('prediction_M13.png')
```
