```python
import ois
from astropy.io.fits import getdata
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
test = 'M13V20120711.fits'
ref = 'M13Vreferencefr.fits'
```

```python
test_img = getdata(test)
ref_img = getdata(ref)
```

```python
diff_img, opt_img, kernel, background = ois.optimal_system(test_img, ref_img)
```
