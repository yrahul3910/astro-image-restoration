### Apply a Gaussian blur

```python
img.shape
```

```python
filter2 = gaussian(img, sigma=4, multichannel=True)
```

### Add random Gaussian noise

```python
noise = np.random.randn(*img.shape)
```

```python
noisy_img = filter2 + noise
```

```python
plt.imshow(noisy_img);
```

```python
noisy_img.shape
```

```python
plt.imsave(fname='noisy_test.png', arr=noisy_img)
```

```python
def add_noise(filename, sigma):
    img = plt.imread(filename)
    fil = gaussian(img, sigma=sigma, multichannel=True)
    noise = np.random.randn(*img.shape)
    
    return fil + noise
```

```python
import os
```

```python
files = os.listdir('VOC2012/JPEGImages')
```

```python
for file in files:
    noisy_img = add_noise('VOC2012/JPEGImages/' + file, sigma=1.6)
    plt.imsave(fname='VOC2012/noisy/' + file.split('.')[0] + '.png', arr=noisy_img)
```

```python
for i, file in enumerate(files):
    print('Processing file ', i + 1)
    img = plt.imread('VOC2012/JPEGImages/' + file)
    blurred = gaussian(img, sigma=4, multichannel=True)
    plt.imsave(fname='VOC2012/gauss_blur/' + file.split('.')[0] + '.png', arr=blurred)
```
