```python
from astropy.io.fits import getdata
from astropy.convolution import AiryDisk2DKernel, convolve
import os
import numpy as np
import pickle
```

```python
files = os.listdir('FITS Files')
```

```python
files
```

```python
kernel = AiryDisk2DKernel(radius=3)
```

```python
for i, file in enumerate(files):
    print('Processing file', (i+1), '\b...', end='')
    data = getdata('FITS Files/' + file).astype(np.float64)
    data /= np.max(data)
    noisy = convolve(data, kernel)
    noisy += np.random.normal(loc=0, scale=0.01)
    
    with open('Noisy/' + file.split('.')[0] + '.pkl', 'wb') as f:
        pickle.dump({'X': data, 'Y': noisy}, f)
    
    print('done!')
```
