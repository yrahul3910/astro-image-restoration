```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdul = fits.open('M13V20130911.fits')
data = hdul[0].data
hdul.close()
hdul = fits.open('M13Vreferencefr.fits')
ref = hdul[0].data
hdul.close()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(data, cmap='gray')
plt.colorbar()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ref, cmap='gray')
plt.colorbar()
```

```python
plt.figure()
plt.imshow(data - ref, cmap='gray')
plt.colorbar()
```
